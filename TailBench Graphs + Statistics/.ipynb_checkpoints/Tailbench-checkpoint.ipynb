```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df=pd.read_csv('./lats-all-stress.txt',delimiter='|')
df.columns=['QueueTimes','ServiceTimes','SojournTimes']
df.head()
```

```python
# A custom function to calculate
# probability distribution function
# def pdf(x):
#     mean = np.mean(x)
#     std = np.std(x)
#     y_out = 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * std**2))
#     return y_out
```

```python
def plotBellCurve(x,message):
    # To generate an array of
    # y-values using corresponding x-values
    mean=np.mean(stats.zscore(x))
    std = np.std(stats.zscore(x))
    y = stats.norm.pdf(stats.zscore(x))


    # Plotting the bell-shaped curve
#     plt.rcParams['agg.path.chunksize'] = 20
    plt.figure(figsize = (30,30))
#     plt.plot(x, y, color = 'black',
#              linestyle = 'solid')
    
    p95=np.percentile(stats.zscore(x),95)
    p99=np.percentile(stats.zscore(x),99)
    
    plt.scatter( stats.zscore(x), y, marker = 'o', s = 5, color ='blue')
    plt.axvline(x=p95,color = 'purple',linestyle='--')
    plt.axvline(x=p99,color = 'brown',linestyle='--')
    plt.axvline(x=mean,color = 'magenta',linestyle='--')

    plt.text(p95+100,0,f"{np.round(p95,2)}",rotation=90)
    plt.text(p99+100,0,f"{np.round(p99,2)}",rotation=90)
    plt.text(mean+100,0,f"{np.round(mean,2)}",rotation=90)

#     plt.fill_between(x, y, 0,alpha = 0.2, color = 'blue')
    plt.xlabel("Latency (in ms)")
    plt.ylabel("Frequency")
    plt.title(f"Normal Probability Distribution For {message}",pad=20)
    plt.savefig("lats-all-stress-QueueTimes.png")
```

```python
plotBellCurve(df["QueueTimes"],"QueueTimes")
```
